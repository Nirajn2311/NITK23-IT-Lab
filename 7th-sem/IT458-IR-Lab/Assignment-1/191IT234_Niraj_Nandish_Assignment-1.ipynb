## Text Preprocessing

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()

def text_lowercase(text):
    return text.lower()

def remove_non_alphanum(text):
    res = ""
    for elem in text:
        if elem.isalnum() or elem == " ":
            res += elem
    return res

def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result

def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

def remove_whitespace(text):
    return  " ".join(text.split())

def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return filtered_text

def stem_words(text):
    word_tokens = word_tokenize(text)
    stems = [stemmer.stem(word) for word in word_tokens]
    return stems

def get_vocab_size():
    unique_words = set()
    for doc in docs:
        for word in doc.split():
            unique_words.add(word)
    return len(unique_words)
```

```python
for i in range(len(docs)):
    docs[i] = remove_numbers(docs[i])

print("Vocabulary size after removing numbers:", str(get_vocab_size()))
```

```python
for i in range(len(docs)):
    docs[i] = remove_whitespace(docs[i])

print("Vocabulary size after removing extra white space:", str(get_vocab_size()))
```

```python
for i in range(len(docs)):
    docs[i] = remove_stopwords(docs[i])

vocab = set()
for doc in docs:
    for word in doc:
        vocab.add(word) 

print("Vocabulary size after removing stop words:", str(len(vocab)))
```

## Inverted Index Construction

```python
inverted_index = {}

for doc in docs:
    for i in range(len(doc)):
        term = doc[i]
        if term in inverted_index:
            res = inverted_index[term]
            res[1].add(i)
            res[0] = len(res[1])
        else: 
            inverted_index[term] = [1,{i}]

sorted(inverted_index.items(), key=lambda x: x[1][0], reverse=True)
```

## Boolean operations

```python
def or_op(set1, set2):
    return set1.union(set2)

def and_op(set1, set2):
    return set1.intersection(set2)

def not_op(set1, set2):
    return set1.difference(set2)
```

## Boolean Retrival

```python
def search_doc (query):
    print("Query:", query)
    final_doc_set = set()
    prev = 'OR'
    for word in query.split():
        if word not in ["AND", "OR", "NOT"]:
            print(word, "->", inverted_index[word])
            cur_term = inverted_index[word][1]
            if prev == 'AND':
                final_doc_set = and_op(final_doc_set, cur_term)
            elif prev == 'OR':
                final_doc_set = or_op(final_doc_set, cur_term)
            elif prev == 'NOT':
                final_doc_set = not_op(final_doc_set, cur_term)
        else:
            prev = word


    if len(final_doc_set) == 0:
        print('Result Set: Empty Set')
    else:
        print('Result Set:', final_doc_set)
```
